# Visualisation of Raypaths using TauP from Obspy

```python
import numpy as np 
import matplotlib.pyplot as plt #for figures
from obspy.taup import TauPyModel #TauP
```

```python
model = TauPyModel(model="prem")
```

```python
parameters = {"source_depth_in_km":55, "distance_in_degree":67, "phase_list":["P", "PP", "PcP", "PKP", "PKiKP"]}

model.get_travel_times(**parameters)
model.get_ray_paths(**parameters)

arrivals = model.get_ray_paths(**parameters)
arrivals.plot()
```

```python
fig = plt.figure(figsize=(20, 10)) 
ax1 = plt.subplot(121) 
ax2 = plt.subplot(122, projection='polar') 
ax2.set_theta_zero_location('N') 
ax2.set_theta_direction(-1)


distances = np.linspace(10, 180, 100)


phases = ["P", "PKP", "PKiKP", "PKIKP", "PP"]
colors = {"P":"b", "PKP":"r", "S":"g", "PKiKP":"k", "PKIKP":"k", "PP":"r"}

source_depth = 55

for i, value in enumerate(distances):
   arrivals = model.get_travel_times(source_depth_in_km=source_depth,
                                     distance_in_degree=value,
                                     phase_list=phases)
   if len(arrivals)==0: continue
   for arrival in arrivals:
      ax1.scatter(value,arrival.time, c=colors[arrival.phase.name], s=10, linewidth=0)
 
ax1.set_ylabel("time (s)")
ax1.set_xlabel("distance (degree)")

```

```python
distances = [60, 90, 120, 150]

fig, ax = plt.subplots(1, 4, figsize=[10,4], subplot_kw={'projection':'polar'})

for i, value in enumerate(distances[::]):
   arrivals = model.get_ray_paths(source_depth_in_km=source_depth,
                                  distance_in_degree=value,
                                  phase_list=phases)
   if len(arrivals)==0: pass
   else:
       arrivals.plot(ax=ax[i], plot_type='spherical', legend=False, label_arrivals=True,
                    show=False)
```
